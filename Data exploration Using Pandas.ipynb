# AGGREGATE, FILTER, TRANSFORM, APPLY
Groupby objects have aggregate(), filter(), transform() and apply() method that efficiency implement a veriaety of useful operation
before combining the grouped data

```python
#for the purpose of the fllg subsections, we will sue this Dataframe
rng = np.random.RandomState(0)
df_gen = pd.DataFrame({"key":['A', 'B', 'C', 'A', 'B','C'], "data":range(6),
    "data2": rng.randint(0,10,6)})
```

```python
df_gen
```

#aggregate can take a string a function or a list thereof , and compute all the aggreagtes
# at once
df_gen.groupby('key').aggregate(['min', np.median, max])

```python
#another useful pattern to pass a dictionary mapping ciolumn names to operation
#ot be applied on that column
df_gen.groupby('key').aggregate({'data':'min', 'data2':max})
```

```python
#a filter() function allows you to drop data based on the group propertice. for example
#me might want to keeps all the groups in which the standard deviation is larger than some criteria value
def filter_func(x):
    return x['data2'].std()>4
print(df_gen.groupby('key').std());
df_gen.groupby('key').filter(filter_func)
```

```python
#transform() return some transformed version of the full data to recombine
#A common exmaple is to center the data by subtractiing the group-wise mean
df_gen.groupby('key').transform(lambda x:x-x.mean())
```

```python
#the apply() method lets you apply an arbituary function to the group results. the function should take a dataframe
#and returbn either a panda objects (e.g, Dataframe, series) or a scaler;
def norma_data(x):
    x['data']/=x['data2'].sum()
    return x
df_gen.groupby('key').apply(norma_data)
```

# DATAFRAME.EVAL() FOR COLUMN-WISE OPERATIONS
just as pandas has a top level pd.eval() function,   dataframe have an eval() method that works in similair ways . the benfits
of the eval() method is that column can be refered to by name.

```python
#we will use this labeled array as an exmaple:
df = pd.DataFrame (rng.rand(1000,3), columns =['A','B','C'])
df.head()
```

```python
#using pd.eval() as above , we cvan operate expression with the three columns like this'\
result1 = (df['A']+df['B'])/ (df['C']-1)
result2 = pd.eval('(df.A+df.B)/(df.C-1)')
np.allclose(result1, result2)
```

```python
#the datafrane.eval() method alow s me=uch more succint evaluation of expression with the columns
result3 = df.eval('(A+B)/(C-1)')
np.allclose(result1, result3)
```

```python
#we can use df.eval() to create a new column 'D' and s=asign to it  a value computed from the pther columns

df.eval('D = (A+B)/C', inplace = True)
df.head()
```

```python
#in the same way , any existing column can be modified
df.eval('D = (A+B)/C', inplace = True)
df.head()
```

```python
#The dataframe.eval() method supporty an additional syntax that let its work witn a local python
#varibles using @ as a reference to the avriable . consider the following
column_mean = df.mean(axis =1)
result1 = df['A']+ column_mean
result2 = df.eval('A+@column_mean')
np.allclose(result1, result2)
```

NOTICE THATS THIS @ CHARACTER IS ONLY SUPPORTED BY THE DATAFRAME.EVAL METHOD NOT BY THE PANDAS.EVAL FUNCTION , BECAUESE THE 
PANDAS.EVAL() FUNCTION ONLY HAS ACCESS TO THE ONE PHTHON NAMESPACE

```python
#CONSIDER THE FOLLOWING
result1 = df[(df.A<0.5) & (df.B<0.5)]
result2 = pd.eval('df[(df.A<0.5) & (df.B<0.5)]')
np.allclose(result1, result2)
```

```python
#this is ana expression invovling columns of the dataframe . it cannot be expressed using the dataframe.eval()
#syntax, howver instead for thuis type of filtering operations , you cann use the query( ) method
result2 = df.query('A<0.5 and B<0.5')
np.allclose(result1, result2)
```

```python
#the query() method also accept the @flag to mark local variables
Cmean = df['C'].mean()
result1 = df[(df.A<Cmean) &(df.B< Cmean)]
result2 = df.query('A<@Cmean and B<@Cmean')
np.allclose(result1, result2)
```
